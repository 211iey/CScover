```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# 예: df는 이미 메모리에 존재한다고 가정 (datetime 포함, series_id 포함, congestion 포함)
df['hour'] = pd.to_datetime(df['datetime']).dt.hour

# 1. 하루 24시간 단위로 시계열 평균 프로파일 생성
pivot = df.pivot_table(index='series_id', columns='hour', values='train_subway21.congestion', aggfunc='mean').fillna(0)

# 2. 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(pivot)

# 3. elbow method로 최적 k 시각화
sse = []
k_range = range(2, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(k_range, sse, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. df 불러오기 (이미 있음이라고 가정)
# datetime은 datetime 형식으로 변환
df['datetime'] = pd.to_datetime(df['datetime'])

# 2. 각 시계열(series_id)에 대해 시간대별 평균 혼잡도 계산 (즉, 0~23시 평균)
df['hour'] = df['datetime'].dt.hour
df_grouped = df.groupby(['series_id', 'hour'])['train_subway21.congestion'].mean().reset_index()

# 3. 피벗: 시계열을 행, 시간(hour)을 열로
pivot_df = df_grouped.pivot(index='series_id', columns='hour', values='train_subway21.congestion').fillna(0)

# 4. 클러스터링
k = 5  # 예시: 클러스터 수 5
kmeans = KMeans(n_clusters=k, random_state=42)
pivot_df['cluster'] = kmeans.fit_predict(pivot_df)

# 5. 클러스터별 평균 시계열 시각화
plt.figure(figsize=(12, 6))
for cluster_id in range(k):
    mean_series = pivot_df[pivot_df['cluster'] == cluster_id].drop(columns='cluster').mean()
    plt.plot(mean_series.index, mean_series.values, label=f'Cluster {cluster_id}')
plt.title("Average Congestion Pattern per Cluster (Hourly)")
plt.xlabel("Hour of Day")
plt.ylabel("Congestion")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances

# 클러스터 수
n_clusters = pivot_df['cluster'].nunique()

# pivot_df에서 series_id는 인덱스로 되어 있음
pivot_df_reset = pivot_df.reset_index()

# 클러스터별 대표 시계열 추출
representatives = []

for cluster_id in range(n_clusters):
    # 해당 클러스터에 속한 시계열만 추출
    cluster_data = pivot_df_reset[pivot_df_reset['cluster'] == cluster_id]

    # 클러스터 중심 계산 (평균 벡터)
    centroid = cluster_data.drop(columns=['series_id', 'cluster']).mean().values.reshape(1, -1)

    # 시계열과 중심 간 거리 계산 (유클리드 거리)
    distances = pairwise_distances(cluster_data.drop(columns=['series_id', 'cluster']), centroid)

    # 가장 가까운 시계열의 index 찾기
    closest_idx = np.argmin(distances)

    # 대표 시계열의 series_id 저장
    representatives.append(cluster_data.iloc[closest_idx]['series_id'])

# 결과 확인
print("대표 시계열 리스트:")
for i, sid in enumerate(representatives):
    print(f"Cluster {i}: {sid}")
```

```python
import pandas as pd
from functools import reduce

# 1. 파일 경로 정의
paths = [
    "/content/drive/MyDrive/25-1/25-1 DScover/DScover_main_기상청/메인C조train_data/메인C조train_data/train_subway21.csv",
    "/content/drive/MyDrive/25-1/25-1 DScover/DScover_main_기상청/메인C조train_data/메인C조train_data/train_subway22.csv",
    "/content/drive/MyDrive/25-1/25-1 DScover/DScover_main_기상청/메인C조train_data/메인C조train_data/train_subway23.csv"
]

# 2. 공통 컬럼 이름 매핑 함수
def standardize_columns(df, prefix):
    df = df.rename(columns={col: col.replace(f"{prefix}.", "") for col in df.columns if col.startswith(prefix)})
    return df

# 3. train 데이터 불러오기 및 전처리
train_dfs = []
for i, path in enumerate(paths, start=21):
    df = pd.read_csv(path)
    df = standardize_columns(df, f"train_subway{i}")
    train_dfs.append(df[[
        "tm", "line", "station_number", "station_name", "direction", "congestion"
    ]])

# 4. train 데이터 병합 및 평균 계산
combined_df = pd.concat(train_dfs)
grouped_df = combined_df.groupby(
    ["tm", "line", "station_number", "station_name", "direction"], as_index=False
)["congestion"].mean()

# 5. test 데이터 불러오기 및 전처리
test_df = pd.read_csv("/content/test_subway.csv")

# test_df에 맞춰 key 컬럼 이름 정리
test_df.rename(columns={
    "TM": "tm",
    "Line": "line",
    "Direction": "direction"
}, inplace=True)

# 6. 병합
merged = pd.merge(
    test_df,
    grouped_df,
    on=["tm", "line", "station_number", "station_name", "direction"],
    how="left"
)

# 7. NaN인 Congestion은 0으로 채움
merged["Congestion"] = merged["congestion"].fillna(0)
merged.drop(columns=["congestion"], inplace=True)  # 중간 계산 컬럼 제거

# 8. 저장
merged.to_csv("/content/final_test_subway_filled.csv", index=False)
```
